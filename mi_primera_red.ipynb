```python
import numpy as np

#definimos las entradas (rojo y redondez)
x= np.array([0.9, 0.8])

#definimos los pesos y el bias
w= np.array([0.7, 0.5])
b= -0.8

#se calcula de suma que es el producto punto
z= np.dot(x,w) + b
# función de activación escalón
y= 1 if z >= 0 else  0

print(f"Resultado de z : {z: .2f}")
print(f"Salida de la neurona y= {y}")
```
