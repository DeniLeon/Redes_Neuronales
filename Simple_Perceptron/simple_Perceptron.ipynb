#### Perceptrón simple para AND

```python

import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

# AND 
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

y = np.array([0,0,0,1], dtype=int)

# Inicialización
w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

# Entrenamiento
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = step(z)
        e = yi - yhat
        
        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1
    
    print(f"Época {epoch+1}: errores={errors}, w={w}, b={b}")
    
    if errors == 0:
        print(f"Convergió en la época {epoch+1}")
        break

print("\nResultados finales:")
print("w =", w)
print("b =", b)

for xi in X:
    z = np.dot(w, xi) + b
    print(xi.astype(int), "->", "   z =", z, step(z))

#Plotting
plt.figure(figsize=(6,6))

# separar por clase
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# dibujar puntos
plt.scatter(X_class0[:,0], X_class0[:,1], s=50, color='blue', label='Clase 0')
plt.scatter(X_class1[:,0], X_class1[:,1], s=50, color='red', label='Clase 1')

# recta de decisión
x_vals = np.linspace(-0.5, 1.5, 200)

if abs(w[1]) > 1e-10:# para evitar división por cero
    y_vals = -(w[0]*x_vals + b)/w[1] #aqui calculamos y en función de x usando la ecuación de la recta w[0]*x + w[1]*y + b = 0
    plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')
elif abs(w[0]) > 1e-10:
    x_line = -b / w[0]
    plt.axvline(x=x_line, color='black', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron - AND Function", color='red')
plt.grid(True)
plt.legend()
plt.show()
```

#### Perceptrón simple para XOR

En la gráfica que se genera a continuación,  se observa que los puntos de las clases están colocados de tal forma que no existe una sola recta que los separe. Esto confirma que  *XOR* no es linealmente separable, por lo que un perceptrón simple no puede resolverlo correctamente.

```python
import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

# XOR truth table
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

y = np.array([0,1,1,0], dtype=int)

# Inicialización
w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

# Entrenamiento
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = step(z)
        e = yi - yhat
        
        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1
    

print("\nResultados finales:")
print("w =", f"{w[0]:.4f}, {w[1]:.4f}")
print("b =", f"{b:.4f}")

for xi in X:
    z = np.dot(w, xi) + b
    print(xi.astype(int), "->", step(z), "   z =", z)

#Plotting
plt.figure(figsize=(6,6))

# separar por clase
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# dibujar puntos
plt.scatter(X_class0[:,0], X_class0[:,1], s=50, color='blue', label='Clase 0')
plt.scatter(X_class1[:,0], X_class1[:,1], s=50, color='red', label='Clase 1')

# recta de decisión
x_vals = np.linspace(-0.5, 1.5, 200)

if abs(w[1]) > 1e-10:
    y_vals = -(w[0]*x_vals + b)/w[1]
    plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')
elif abs(w[0]) > 1e-10:
    x_line = -b / w[0]
    plt.axvline(x=x_line, color='black', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron - XOR Function",color = 'green')
plt.grid(True)
plt.legend()
plt.show()
```

#### Perceptrón Simple: Fitting de una Recta

En esta sección se adapta la arquitectura del **perceptrón simple** para intentar realizar una regresión lineal. Para ello, se modifica la función de activación de escalón por una **función de activación lineal**:

$$f(z) = z$$

#### Limitantes del Modelo

El modelo presenta las siguientes limitantes técnicas en tareas de *fitting*:

* [cite_start]**Naturaleza del algoritmo:** El algoritmo de aprendizaje del perceptrón simple está diseñado originalmente para problemas de clasificación, donde la salida del error toma valores discretos de $0, 1, -1$.
* **Precisión en datos continuos:** Al aplicar este modelo a datos continuos, el error (la diferencia entre el valor de predicción y el valor real) casi nunca es exactamente cero debido a los decimales involucrados.
* **Falta de convergencia:** El contador de errores nunca llega a cero; por lo tanto, no se alcanza la condición de convergencia y el algoritmo continúa ajustando los pesos indefinidamente o hasta agotar las épocas establecidas.

> **Conclusión:** Lo anterior demuestra que el modelo de Perceptrón Simple **no es óptimo** para tareas de *fitting*.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Datos de una recta (y = 2x + 1)
X = np.array([[1], [2], [3], [4], [5]], dtype=float)
y = np.array([3.1, 4.9, 7.2, 8.8, 11.1], dtype=float)

# 2. Inicialización (Igual que en tu ejemplo de AND)
w = np.zeros(1, dtype=float)
b = 0.0
lr = 0.01
epochs = 50

# 3. Entrenamiento con lógica de Perceptrón
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = z  # Activación lineal para intentar el fitting
        e = yi - yhat
        
        # El Perceptrón "salta" con cualquier diferencia, por mínima que sea
        if e!= 0: 
            w += lr * e * xi
            b += lr * e
            errors += 1
    
    if errors == 0:
        print(f"Convergió en época {epoch}")
        break


print(f"Resultado final: w={w[0]:.2f}, b={b:.2f}")
# 4. Visualización
plt.scatter(X, y, color='red', label='Datos reales')
#x_range es un array de 100 puntos entre 0 y 6 para dibujar la recta ajustada
x_range = np.linspace(0, 6, 100)
#w[0]*x + b es la ecuación de la recta ajustada por el Perceptrón, W[0] porque w es un array de un solo elemento
plt.plot(x_range, w[0]*x_range + b, color='blue', label='Recta Perceptrón')
plt.title("Perceptrón simple fitting a una  recta")
plt.legend()
plt.grid(True)
plt.show()
```
