```python
import numpy as np

def step(z):
    return 1 if z >= 0 else 0

# AND truth table
X = np.array([[0,0],[0,1],[1,0],[1,1]], dtype=float)
y = np.array([0,0,0,1], dtype=int)  

w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        yhat = step(np.dot(w, xi) + b)
        e = yi - yhat         
        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1
    if errors == 0:
        break

print("w =", w, "b =", b)
for xi in X:
    print(xi.astype(int), "->", step(np.dot(w, xi) + b))
```
