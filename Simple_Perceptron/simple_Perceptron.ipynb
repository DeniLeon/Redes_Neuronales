**Perceptrón simple para AND**

```python

import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

# AND 
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

y = np.array([0,0,0,1], dtype=int)

# Inicialización
w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

# Entrenamiento
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = step(z)
        e = yi - yhat
        
        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1
    
    print(f"Época {epoch+1}: errores={errors}, w={w}, b={b}")
    
    if errors == 0:
        print(f"Convergió en la época {epoch+1}")
        break

print("\nResultados finales:")
print("w =", w)
print("b =", b)

for xi in X:
    z = np.dot(w, xi) + b
    print(xi.astype(int), "->", step(z), "   z =", z)

#Plotting
plt.figure(figsize=(6,6))

# separar por clase
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# dibujar puntos
plt.scatter(X_class0[:,0], X_class0[:,1], s=50, color='blue', label='Clase 0')
plt.scatter(X_class1[:,0], X_class1[:,1], s=50, color='red', label='Clase 1')

# recta de decisión
x_vals = np.linspace(-0.5, 1.5, 200)

if abs(w[1]) > 1e-10:# para evitar división por cero
    y_vals = -(w[0]*x_vals + b)/w[1] #aqui calculamos y en función de x usando la ecuación de la recta w[0]*x + w[1]*y + b = 0
    plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')
elif abs(w[0]) > 1e-10:
    x_line = -b / w[0]
    plt.axvline(x=x_line, color='black', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron - AND Function", color='red')
plt.grid(True)
plt.legend()
plt.show()
```

**Perceptrón simple para OR**

```python
import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

# OR truth table
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

y = np.array([0,1,1,1], dtype=int)

# Inicialización
w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

# Entrenamiento
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = step(z)
        e = yi - yhat

        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1

    print(f"Época {epoch+1}: errores={errors}, w={w}, b={b}")

    if errors == 0:
        print(f"Convergió en la época {epoch+1}")
        break

print("\nResultados finales:")
print("w =", w)
print("b =", b)

for xi in X:
    z = np.dot(w, xi) + b
    print(xi.astype(int), "->", step(z), "   z =", z)

#Plotting
plt.figure(figsize=(6,6))

# separar por clase
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# dibujar puntos
plt.scatter(X_class0[:,0], X_class0[:,1], s=50, color='blue', label='Clase 0')
plt.scatter(X_class1[:,0], X_class1[:,1], s=50, color='red', label='Clase 1')

# recta de decisión
x_vals = np.linspace(-0.5, 1.5, 200)

if abs(w[1]) > 1e-10:
    y_vals = -(w[0]*x_vals + b)/w[1]
    plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')
elif abs(w[0]) > 1e-10:
    x_line = -b / w[0]
    plt.axvline(x=x_line, color='black', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron - OR Function")
plt.grid(True)
plt.legend()
plt.show()
```

**Perceptrón simple para XOR**
En la gráfica que se genera a continuación,  se observa que los puntos de las clases están colocados de tal forma que no existe una sola recta que los separe. Esto confirma que  XOR no es linealmente separable, por lo que un perceptrón simple no puede resolverlo correctamente.

```python
import numpy as np
import matplotlib.pyplot as plt

def step(z):
    return 1 if z >= 0 else 0

# XOR truth table
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]], dtype=float)

y = np.array([0,1,1,0], dtype=int)

# Inicialización
w = np.zeros(2, dtype=float)
b = 0.0
lr = 0.1
epochs = 50

# Entrenamiento
for epoch in range(epochs):
    errors = 0
    for xi, yi in zip(X, y):
        z = np.dot(w, xi) + b
        yhat = step(z)
        e = yi - yhat
        
        if e != 0:
            w += lr * e * xi
            b += lr * e
            errors += 1
    
    print(f"Época {epoch+1}: errores={errors}, w={w}, b={b}")

print("\nResultados finales:")
print("w =", w)
print("b =", b)

for xi in X:
    z = np.dot(w, xi) + b
    print(xi.astype(int), "->", step(z), "   z =", z)

#Plotting
plt.figure(figsize=(6,6))

# separar por clase
X_class0 = X[y == 0]
X_class1 = X[y == 1]

# dibujar puntos
plt.scatter(X_class0[:,0], X_class0[:,1], s=50, color='blue', label='Clase 0')
plt.scatter(X_class1[:,0], X_class1[:,1], s=50, color='red', label='Clase 1')

# recta de decisión
x_vals = np.linspace(-0.5, 1.5, 200)

if abs(w[1]) > 1e-10:
    y_vals = -(w[0]*x_vals + b)/w[1]
    plt.plot(x_vals, y_vals, color='black', label='Decision Boundary')
elif abs(w[0]) > 1e-10:
    x_line = -b / w[0]
    plt.axvline(x=x_line, color='black', label='Decision Boundary')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron - XOR Function",color = 'green')
plt.grid(True)
plt.legend()
plt.show()
```

**fitting a una recta**

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([0, 1, 2, 3, 4, 5], dtype=float)
y = np.array([1, 2, 5, 8, 8, 11], dtype=float)   # y = 2x + 1

# Inicialización
w = 0.0
b = 0.0
lr = 0.01
epochs = 50
n = len(X)

# Para ver cómo evoluciona el error
loss_history = []

#Entrenamiento 
for epoch in range(epochs):
    yhat = w * X + b
    error = yhat - y
    
    mse = np.mean(error**2)
    #print("mse:",mse)
    loss_history.append(mse)
    
    # gradientes
    dw = (2/n) * np.sum(error * X)
    db = (2/n) * np.sum(error)
    
    # actualización
    w -= lr * dw
    b -= lr * db

#Resultados
print("Fitting a una recta")
print("w =", w)
print("b =", b)

for i, mse in enumerate(loss_history):
    print(f"Epoch {i+1}: MSE = {mse}")

print("\nPredicciones:")
for xi, yi_real in zip(X, y):
    yi_pred = w * xi + b
    print(f"x = {xi:.1f}, y real = {yi_real:.2f}, y pred = {yi_pred:.2f}")

#plotting
plt.figure(figsize=(6,6))

# puntos
plt.scatter(X, y, s=50, color='blue')

# recta ajustada
x_vals = np.linspace(min(X)-1, max(X)+1, 200)
y_vals = w * x_vals + b

plt.plot(x_vals, y_vals, color='red', label='Recta ajustada')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Fitting a una recta")
plt.grid(True)
plt.legend()
plt.show()
```
